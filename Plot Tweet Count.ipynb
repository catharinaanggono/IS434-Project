```python
import pandas as pd

# load in the csv file
tweets = pd.read_csv('./Final Dataset/2022 Twitter Tweets Jan-April_Amber Heard Data.csv')

# create a dataframe
df = pd.DataFrame(tweets)

# filter the dataframe to contain just english tweets
df = df[df["lang"] == "en"]

# convert 'id_str' column values to Int64
df = df.astype({'id_str':'int64'})

# convert 'id_str' column values to float then to Int64
df['in_reply_to_status_id_str'] = df['in_reply_to_status_id_str'].astype('float').astype('Int64')
# df = df[(df['created_at'] > '2022-01-01') & (df['created_at'] < '2022-02-01')]

tweets = pd.read_excel('./Final Dataset/DURING 11-30 Apr Twitter Data.xlsx')
df_apr2 = pd.DataFrame(tweets)

tweets = pd.read_excel('./Final Dataset/DURING 1May - 1Jun Twitter Data.xlsx')
df_may = pd.DataFrame(tweets)

tweets = pd.read_excel('./Final Dataset/POST 2-30 Jun Twitter Data.xlsx')
df_june = pd.DataFrame(tweets)

tweets = pd.read_excel('./Final Dataset/POST 1-31 Jul Twitter Data.xlsx')
df_july = pd.DataFrame(tweets)

tweets = pd.read_excel('./Final Dataset/POST 1-31 Aug Twitter Data.xlsx')
df_aug = pd.DataFrame(tweets)
```

```python
df_jan = df[(df['created_at'] >= '2022-01-01') & (df['created_at'] < '2022-02-01')]
df_jan

df_feb = df[(df['created_at'] >= '2022-02-01') & (df['created_at'] < '2022-03-01')]
df_feb

df_mar = df[(df['created_at'] >= '2022-03-01') & (df['created_at'] < '2022-04-01')]
df_mar

df_apr = df[(df['created_at'] >= '2022-04-01') & (df['created_at'] < '2022-04-11')]
df_apr

df_may = df_may[(df_may['created_at'] >= '2022-05-01') & (df_may['created_at'] < '2022-06-01')]
df_may

df_june = df_june.append(df_may[(df_may['created_at'] > '2022-05-31') & (df_may['created_at'] < '2022-06-02')])
df_june
```

```python
import matplotlib.pyplot as plt

# month_list = df['month'].tolist()
# unique_month = list(set(month_list))
month = {1: "January", 2: "February", 3: "March", 4: "April", 5: "May", 6: "June", 7: "July", 8: "August", 9: "September", 10: "October", 11: "November", 12: "December"}

x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug']
y = [len(df_jan.index), len(df_feb.index), len(df_mar.index), len(df_apr.index) + len(df_apr2.index), len(df_may.index), len(df_june.index), len(df_july.index), len(df_aug.index) ]

# print(month_list)
# Show graph
plt.figure(figsize=(16, 8))

plt.plot(x, y)

plt.title("Amber Heard Tweet Count Trend")
plt.xlabel("Months (2022)")
plt.ylabel("Count")
plt.show()
```
