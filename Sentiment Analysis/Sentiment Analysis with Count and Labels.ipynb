```python
import pandas as pd
import spacy
from spacy import displacy
from spacy import tokenizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Load the NER model
nlp = spacy.load("en_core_web_sm")

# Load the sentiment analyzer
sid = SentimentIntensityAnalyzer()
```

```python
#reading pre trial data set
df = pd.read_csv('PRE 2022 Twitter Tweets Jan-April_Amber Heard Data.csv')
```

```python
#getting english tweets from the pre trial data set
df_en = df[df["lang"] == "en"]
df_en
```

```python
#making a data frame from the january tweets 
df_jan = df_en[(df_en['created_at'] > '2022-01-01') & (df_en['created_at'] < '2022-02-01')]
df_jan
```

```python
#getting sentiment count in jan and who it is directed towards

# Define a function to extract named entities and classify sentiment for a tweet
def extract_sentiment(text, entity):
    sentiment_scores = sid.polarity_scores(text)
    if entity.lower() in text.lower():
        if sentiment_scores['compound'] > 0:
            return 'positive'
        elif sentiment_scores['compound'] < 0:
            return 'negative'
        else:
            return 'neutral'
    else:
        return None

# Apply named entity recognition and sentiment analysis to each tweet in the dataframe
sentiment_cols = []
for entity in ['Johnny Depp', 'Amber Heard']:
    df_jan[f"{entity} sentiment"] = df_jan['full_text'].apply(lambda x: extract_sentiment(x, entity))
    sentiment_cols.append(f"{entity} sentiment")

# Group by the index of the dataframe and sum the sentiment counts for each entity
sentiment_counts = df_jan.groupby(df_jan.index)[sentiment_cols].sum()

# Print the sentiment counts for Johnny Depp and Amber Heard
depp_sentiment_counts = sentiment_counts['Johnny Depp sentiment'].value_counts().drop(0)
heard_sentiment_counts = sentiment_counts['Amber Heard sentiment'].value_counts().drop(0)
print(f"Sentiment counts for Johnny Depp:\n{depp_sentiment_counts}\n")
print(f"Sentiment counts for Amber Heard:\n{heard_sentiment_counts}")

```
