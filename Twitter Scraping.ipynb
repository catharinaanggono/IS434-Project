```python
import datetime
import pandas as pd
import snscrape.modules.twitter as sntwitter

# Define the query and limit
query = "Amber Heard"
limit = 4999

# Define the date range - scraping daily (during trial: April 11 to June 1 2022, post trial: June 2 to August 31 2022)
start_date = datetime.date(2022, 4, 11)
end_date = datetime.date(2022, 4, 12)

# Create an empty list to store the tweets
tweets = []

# Scrape the tweets   
for tweet in sntwitter.TwitterSearchScraper(query + f" since:{start_date} until:{end_date} lang:en").get_items():
    if len(tweets) > limit:
        break
    else:
        tweets.append([tweet.date, tweet.id, tweet.user.username, tweet.user.id, tweet.inReplyToTweetId, tweet.inReplyToUser, tweet.likeCount, tweet.content])

# Convert the list of tweets to a pandas DataFrame
df = pd.DataFrame(tweets, columns=['created_at', 'id_str', 'user_name', 'user_id', 'in_reply_to_status_id_str', "in_reply_to_username", "like_count",'full_text'])

# Converting the username from the profile link to only its username
df['in_reply_to_username'] = df['in_reply_to_username'].apply(lambda x: str(x).split('/')[-1] if pd.notnull(x) else x)
```
