```python
import pandas as pd

# load in the csv file
tweets = pd.read_csv('./Labelled Dataset/jan_labelled.csv')

# create a dataframe
df = pd.DataFrame(tweets)
df.rename(columns={'user.screen_name': 'user_name', 'user.id_str': 'user_id', 'Johnny Depp sentiment': 'jd_sentiment', 'Amber Heard sentiment': 'ah_sentiment'}, inplace=True)

tweets = pd.read_csv('./Labelled Dataset/feb_labelled.csv')
df_feb = pd.DataFrame(tweets)
df_feb.rename(columns={'user.screen_name': 'user_name', 'user.id_str': 'user_id'}, inplace=True)

tweets = pd.read_csv('./Labelled Dataset/mar_labelled.csv')
df_mar = pd.DataFrame(tweets)
df_mar.rename(columns={'user.screen_name': 'user_name', 'user.id_str': 'user_id'}, inplace=True)

tweets = pd.read_csv('./Labelled Dataset/pre_apr_labelled.csv')
df_apr = pd.DataFrame(tweets)
df_apr.rename(columns={'user.screen_name': 'user_name', 'user.id_str': 'user_id'}, inplace=True)

tweets = pd.read_csv('./Labelled Dataset/apr_labelled.csv')
df_apr2 = pd.DataFrame(tweets)

tweets = pd.read_csv('./Labelled Dataset/may_labelled.csv')
df_may = pd.DataFrame(tweets)

tweets = pd.read_csv('./Labelled Dataset/jun_labelled.csv')
df_june = pd.DataFrame(tweets)

tweets = pd.read_csv('./Labelled Dataset/jul_labelled.csv')
df_july = pd.DataFrame(tweets)

tweets = pd.read_csv('./Labelled Dataset/aug_labelled.csv')
df_aug = pd.DataFrame(tweets)
```

```python
df = df.append(df_feb)
df = df.append(df_mar)
df = df.append(df_apr)
df = df.append(df_apr2)
df = df.append(df_may)
df = df.append(df_june)
df = df.append(df_july)
df = df.append(df_aug)
```

```python
df
```

```python
# get the frequency count of tweet id

tweet_id_with_frequency = df['in_reply_to_status_id_str'].value_counts().to_frame()
tweet_id_with_frequency
```

```python
# drop the ids with count less than 30

tweet_id_with_frequency.drop(tweet_id_with_frequency[tweet_id_with_frequency['in_reply_to_status_id_str'] <= 30].index, inplace = True)
tweet_id_with_frequency
```

```python
# create a list from the tweet ids

tweet_id_with_frequency_list = tweet_id_with_frequency.in_reply_to_status_id_str.keys().tolist()
```

```python
tweet_id_with_frequency_list
```

```python
high_centrality_df = df[df['in_reply_to_status_id_str'].isin(tweet_id_with_frequency_list)]
```

```python
high_centrality_df
```

```python
high_centrality_df.to_csv("Master_High_Centrality_Dataset.csv", index=False, encoding='utf-8')
```
